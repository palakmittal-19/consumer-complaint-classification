```python
# Load the CSV file into a Spark DataFrame
df = spark.read.csv("/content/complaints_processed.csv", header=True, inferSchema=True)

# Show schema and first few rows
df.printSchema()
df.show(5, truncate=False)
```

```python
from pyspark.sql.functions import col, lower, regexp_replace, trim


df = df.drop("Unnamed: 0")


df = df.withColumnRenamed("product", "label") \
       .withColumnRenamed("narrative", "text")


df = df.dropna(subset=["label", "text"])


df = df.withColumn("text_clean", lower(col("text")))
df = df.withColumn("text_clean", regexp_replace(col("text_clean"), "[^a-zA-Z0-9\\s]", " "))
df = df.withColumn("text_clean", regexp_replace(col("text_clean"), "\\s+", " "))
df = df.withColumn("text_clean", trim(col("text_clean")))


df.select("label", "text_clean").show(10, truncate=False)
```

```python
# Count complaints per category
class_dist = df.groupBy("label").count().orderBy("count", ascending=False)
class_dist.show(20, truncate=False)
```

```python
from pyspark.sql.functions import length

# Add a column for text length
df = df.withColumn("text_length", length(col("text_clean")))

# Show basic statistics
df.select("text_length").describe().show()
```

```python
df.select("label", "text_clean").show(10, truncate=False)
```

```python
from pyspark.sql.functions import length, col
import matplotlib.pyplot as plt

# 1) Class distribution using PySpark
class_dist = (
    df.groupBy("label")
      .count()
      .orderBy("count", ascending=False)
)

# Show the PySpark summary table
class_dist.show(truncate=False)

# 2) Convert only this small aggregated result to pandas for plotting
class_dist_pd = class_dist.toPandas()

# 3) Plot using matplotlib
plt.figure(figsize=(10, 5))
plt.bar(class_dist_pd["label"], class_dist_pd["count"])
plt.xticks(rotation=45, ha='right')
plt.title("Class Distribution of Complaint Categories (PySpark Aggregation)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
from pyspark.sql.functions import length, col, when

# 1) Create a text length column (already done earlier, but we re-run to be safe)
df = df.withColumn("text_length", length(col("text_clean")))

# 2) Bucket text lengths into ranges using PySpark
df_binned = df.withColumn(
    "length_bucket",
    when(col("text_length") < 50, "<50")
    .when((col("text_length") >= 50) & (col("text_length") < 100), "50-100")
    .when((col("text_length") >= 100) & (col("text_length") < 200), "100-200")
    .when((col("text_length") >= 200) & (col("text_length") < 400), "200-400")
    .when((col("text_length") >= 400) & (col("text_length") < 800), "400-800")
    .otherwise("800+")
)

# 3) Count how many texts fall into each bucket (Pure PySpark)
length_dist = (
    df_binned.groupBy("length_bucket")
             .count()
             .orderBy("count", ascending=False)
)

length_dist.show()
```

```python
import matplotlib.pyplot as plt

length_pd = length_dist.toPandas()

plt.figure(figsize=(8,5))
plt.bar(length_pd["length_bucket"], length_pd["count"])
plt.title("Text Length Distribution (PySpark Aggregation)")
plt.xlabel("Text Length Bucket")
plt.ylabel("Number of Complaints")
plt.show()
```

```python
train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)

print("Train size:", train_df.count())
print("Test size:", test_df.count())
```

```python
from pyspark.ml.feature import Tokenizer, StopWordsRemover, CountVectorizer, IDF, StringIndexer, PCA, StandardScaler
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, LinearSVC, OneVsRest
from pyspark.ml import Pipeline

# 1. Label indexer
label_indexer = StringIndexer(inputCol="label", outputCol="label_index")

# 2. Tokenizer
tokenizer = Tokenizer(inputCol="text_clean", outputCol="words")

# 3. Stopwords remover
remover = StopWordsRemover(inputCol="words", outputCol="filtered_words")

# 4. CountVectorizer (safe but richer)
cv = CountVectorizer(
    inputCol="filtered_words",
    outputCol="raw_features",
    vocabSize=1500,   # SAFE BUT RICHER
    minDF=8
)

# 5. TF-IDF
idf = IDF(inputCol="raw_features", outputCol="features")

# 6. PCA (safe range)
pca = PCA(
    k=50,                     # SAFE + MORE INFORMATION
    inputCol="features",
    outputCol="pcaFeatures"
)

# 7. Standard Scaler
scaler = StandardScaler(
    inputCol="pcaFeatures",
    outputCol="scaledFeatures",
    withMean=True,
    withStd=True
)
```

```python
# Logistic Regression
lr = LogisticRegression(
    featuresCol="scaledFeatures",
    labelCol="label_index",
    maxIter=40,
    regParam=0.02
)

# Random Forest
rf = RandomForestClassifier(
    featuresCol="scaledFeatures",
    labelCol="label_index",
    numTrees=80,
    maxDepth=12,
    subsamplingRate=0.8
)

# Linear SVM (needs One-vs-Rest for multiclass)
svm = LinearSVC(
    featuresCol="scaledFeatures",
    labelCol="label_index",
    maxIter=40,
    regParam=0.01
)

svm_ovr = OneVsRest(classifier=svm,
                    featuresCol="scaledFeatures",
                    labelCol="label_index")
```

```python
pipeline_lr = Pipeline(stages=[label_indexer, tokenizer, remover, cv, idf, pca, scaler, lr])
pipeline_rf = Pipeline(stages=[label_indexer, tokenizer, remover, cv, idf, pca, scaler, rf])
pipeline_svm = Pipeline(stages=[label_indexer, tokenizer, remover, cv, idf, pca, scaler, svm_ovr])
```

```python
train_small = train_df.sample(False, 0.12, seed=42)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

evaluator = MulticlassClassificationEvaluator(
    labelCol="label_index",
    predictionCol="prediction",
    metricName="accuracy"
)

def train_and_eval(pipeline, name):
    model = pipeline.fit(train_small)
    preds = model.transform(test_df)
    acc = evaluator.evaluate(preds)
    print(f"{name} Accuracy:", acc)
    return acc

acc_lr = train_and_eval(pipeline_lr, "Logistic Regression")
acc_rf = train_and_eval(pipeline_rf, "Random Forest")
acc_svm = train_and_eval(pipeline_svm, "SVM-OVR")
```
